## Matplotlib回顾

例如：使用matplotlib的plot函数绘制sin函数,并添加标题、轴标签等信息。
x = np.arange(0, 10, 0.1)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title('Sin 图像')
plt.xlabel('X-轴')
plt.ylabel('Y-轴')
plt.show()
```

## 小试牛刀——1. 练习数据访问和数据修改
创建series对象并赋值[2, 4, 6, 8, 10]，索引依次是'ABCDE' 
1. 获取series对象的所有索引
2. 获取series对象的第三个索引（'C'）的值
3. 获取series对象的第二到第五的索引的值
4. 判断索引是否包含索引'D'
5. 修改某个索引对应的值
6. 添加一个新的索引值 'F' （value = 100）到series对象中

```python
import pandas as pd

data = [2, 4, 6, 8, 10]
index = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=index)
all_indexes = series.index
print("获取series对象的所有索引:", all_indexes)
value_at_c = series['C']
print("获取series对象的第三个索引('C')的值:", value_at_c)
values_from_2_to_5 = series['B':'E']
print("获取series对象的第二到第五的索引的值:\n", values_from_2_to_5)
contains_d = 'D' in series.index
print("判断索引是否包含索引'D':", contains_d)
series['D'] = 12
print("修改某个索引对应的值:\n", series)
series['F'] = 100
print("添加一个新的索引值 'F' (value = 100)到series对象中:\n", series)
```

## 2. 数据访问与修改练习
创建一个Series对象并赋值[15, 30, 45, 60, 75]，索引依次是'Mon', 'Tue', 'Wed', 'Thu', 'Fri'。  
1. 获取Series对象的第一个索引的值。  
2. 获取Series对象的最后一个索引的值。  
3. 获取索引为'Tue'的值。
4. 获取索引为'Wed'和'Fri'的值。
5. 判断索引为'Sat'是否存在于Series中。
6. 将索引为'Thu'的值修改为90。
7. 将新的索引值 'Sat'（value = 100）添加到Series对象中。

```python
import pandas as pd

data = [15, 30, 45, 60, 75]
index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
series = pd.Series(data, index=index)

first_value = series.iloc[0]
print("获取Series对象的第一个索引的值:", first_value)

last_value = series.iloc[-1]
print("获取Series对象的最后一个索引的值:", last_value)

tue_value = series['Tue']
print("获取索引为'Tue'的值:", tue_value)

wed_and_fri_values = series[['Wed', 'Fri']]
print("获取索引为'Wed'和'Fri'的值:\n", wed_and_fri_values)

contains_sat = 'Sat' in series.index
print("判断索引为'Sat'是否存在于Series中':", contains_sat)

series['Thu'] = 90
print("将索引为'Thu'的值修改为90:\n", series)

series['Sat'] = 100
print("将新的索引值 'Sat'（value = 100）添加到Series对象中\n", series)
```
