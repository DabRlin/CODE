```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 
x = np.arange(-4.0, 4.0, 0.25)
y = np.arange(-3.0, 3.0, 0.25)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2

# 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  #  3D subplot
surf = ax.plot_surface(X, Y, Z, cmap='rainbow')  # 
fig.colorbar(surf)  # 
ax.set_title('Surface Plot')  # 
ax.set_xlabel('X axis')  #  x 
ax.set_ylabel('Y axis')  #  y 
ax.set_zlabel('Z axis')  #  z 
plt.show()
```

```python
import matplotlib.pyplot as plot

from mpl_toolkits.mplot3d import axes3d

from matplotlib import cm

import numpy as np

x1=np.arange(-5,5,0.25)

y1=np.arange(-5,5,0.25)

x1,y1=np.meshgrid(x1,y1)

r1=np.sqrt(x1**2+y1**2)

z1=np.sin(r1)

fig=plt.figure()

ax=fig.add_subplot(111,projection='3d')

ax.plot_surface(x1,y1,z1,cmap=cm.coolwarm,linewidth=0,

                   antialiased=False)

ax.set_zlim(-1.01,1.01)

plt.title('39')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
data = "tttttssttsstttsstsssstststsstsssstststssssstssssstsstsssstsssssttsttssttttstttssttttttttt"

# 
fig, ax = plt.subplots(figsize=(9, 9))

# 
x, y = 0, 0

# 
for c in data:
    if c == "s":
        x -= 1
    elif c == "t":
        x += 1
    ax.plot([x, x], [y, y + 0.1], linewidth=0.75)
    y += 0.1

# 
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("st")

# 
plt.show()
```
