**练习题5 - 数据分析与数据清洗：**

给定一个包含销售订单数据的 DataFrame，请完成以下任务：

```python
data = {'Order_ID': [101, 102, 103, 104, 105],
        'Product_Price': [25.5, 30.0, 22.0, 25.5, 28.0],
        'Quantity': [100, 120, 80, 105, 95]}
df = pd.DataFrame(data)

任务：
- 找到每个订单的总金额（Product_Price * Quantity）.
- 找到总金额最高的订单号和总金额。
```

````python
#练习题1
import pandas as pd

data = {'Subject1': [85, 90, 78, 92, 88],
        'Subject2': [76, 82, 79, 95, 88],
        'Subject3': [88, 90, 92, 80, 76]}
df = pd.DataFrame(data)

# 计算每个科目的平均分
mean_scores = df.mean()

# 找到每个科目的最高分和最低分
max_scores = df.max()
min_scores = df.min()

# 找到所有学生的总分
df['Total'] = df.sum(axis=1)

# 找到总分最高的学生姓名和总分
max_total_student = df[df['Total'] == df['Total'].max()]
max_total_student_name = max_total_student.index[0]
max_total_score = max_total_student['Total'].values[0]

print("平均分:\n", mean_scores)
print("每个科目的最高分:\n", max_scores)
print("每个科目的最低分:\n", min_scores)
print("每个学生的总分:\n", df)
print("总分最高的学生姓名:", max_total_student_name)
print("总分最高的学生总分:", max_total_score)
````

````python
#练习题3
import pandas as pd
data = {'Stock1': [50, 55, 60, 65, 70],
        'Stock2': [45, 50, 55, 60, 65]}
df = pd.DataFrame(data)

# 计算每只股票的第25百分位数和第75百分位数
quantile_25 = df.quantile(0.25)
quantile_75 = df.quantile(0.75)
print("每只股票的第25百分位数:")
print(quantile_25)
print("\n每只股票的第75百分位数:")
print(quantile_75)

# 找到每只股票的中位数
median = df.median()
print("\n每只股票的中位数:")
print(median)
````

````python
#练习题4
import pandas as pd
data = {'City': ['A', 'B', 'C', 'D', 'E'],
        'Max_Temperature': [28, 32, 30, 28, 32],
        'Min_Temperature': [15, 18, 20, 15, 17]}
df = pd.DataFrame(data)

# 找到每个城市的最高温度和最低温度
max_temperature = df[['City', 'Max_Temperature']]
min_temperature = df[['City', 'Min_Temperature']]
print("每个城市的最高温度:")
print(max_temperature)
print("\n每个城市的最低温度:")
print(min_temperature)

# 找到每个城市的众数温度
mode_max_temperature = df['Max_Temperature'].mode().values[0]
mode_min_temperature = df['Min_Temperature'].mode().values[0]
print("\n每个城市的众数温度:")
print("最高温度众数:", mode_max_temperature)
print("最低温度众数:", mode_min_temperature)
````

````python
#练习题5
import pandas as pd
data = {'Order_ID': [101, 102, 103, 104, 105],
        'Product_Price': [25.5, 30.0, 22.0, 25.5, 28.0],
        'Quantity': [100, 120, 80, 105, 95]}
df = pd.DataFrame(data)

# 找到每个订单的总金额（Product_Price * Quantity）
df['Total_Amount'] = df['Product_Price'] * df['Quantity']
print("每个订单的总金额:")
print(df[['Order_ID', 'Total_Amount']] )

# 找到总金额最高的订单号和总金额
max_total_order = df[df['Total_Amount'] == df['Total_Amount'].max()]
print("\n总金额最高的订单号和总金额:")
print(max_total_order[['Order_ID', 'Total_Amount']])
````
