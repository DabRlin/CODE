3. 数据清理

描述对数据进行的清理步骤，包括处理缺失值、异常值、删除重复项、数据类型转换等。

```python
import pandas as pd

# 读取CSV文件
file_path = r'C:\Users\dangh\code\Python\heart.csv'
data = pd.read_csv(file_path)

# 定义列名
columns = ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall', 'output']
data.columns = columns

# 删除含有小于0的行
for column in columns:
    if column != 'output':  # 排除目标列
        data = data[data[column] >= 0]

# 处理缺失值（示例：用平均值填充）
data.fillna(data.mean(), inplace=True)

# 删除重复项
data.drop_duplicates(inplace=True)
```

4. 数据分析
使用Pandas、NumPy等工具，进行数据分析的具体步骤。可以包括计算统计量、创建新特征、分组统计等操作，并提取关键信息和特征，描述数据分布和趋势，分析变量之间的关系和影响等等。描述你对数据的各种分析，例如计算平均值、标准差、频率分布等。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# 读取CSV文件
file_path = r'C:\Users\dangh\code\Python\heart.csv'
data = pd.read_csv(file_path)

# 定义列名和字段描述
columns = {
    'age': '患者年龄',
    'sex': '患者性别',
    'cp': '胸痛类型',
    'trtbps': '静息血压（mm Hg）',
    'chol': '胆固醇（mg/dl）',
    'fbs': '空腹血糖（>120 mg/dl）',
    'restecg': '静息心电图结果',
    'thalachh': '最大心率达到',
    'exng': '运动诱发心绞痛',
    'oldpeak': '运动引起的ST段相对静息时的压低',
    'slp': '峰值运动ST段的斜率',
    'caa': '主要血管数',
    'thall': 'Tl应激测试结果',
    'output': '心脏病发作几率（0=更少，1=更多）'
}

# 修改列名为字段描述
data.columns = [columns[col] for col in data.columns]

# 描述性统计分析
mean_values = data.mean()
std_values = data.std()
value_counts = data['胸痛类型'].value_counts()

# 创建新特征
data['年龄_性别'] = data['患者年龄'] * data['患者性别']

# 分组统计
grouped_data = data.groupby('静息心电图结果').mean()

# 数据分布和趋势分析
plt.hist(data['最大心率达到'])
plt.xlabel('最大心率达到')
plt.ylabel('频率')
plt.title('最大心率达到的直方图')
plt.show()

sns.boxplot(x=data['胆固醇（mg/dl）'], y=data['患者性别'])
plt.ylabel('患者性别')
plt.title('胆固醇水平按性别划分的箱线图')
plt.show()

corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('相关性热力图')
plt.show()

sns.pairplot(data[['患者年龄', '胆固醇（mg/dl）', '最大心率达到', '胸痛类型', '心脏病发作几率（0=更少，1=更多）']], hue='心脏病发作几率（0=更少，1=更多）')
plt.title('所选特征的成对关系图')
plt.show()

grouped = data.groupby(['胸痛类型', '心脏病发作几率（0=更少，1=更多）']).size().unstack()
grouped.plot(kind='bar', stacked=True)
plt.title('胸痛类型与心脏病发作几率的堆叠柱状图')
plt.xlabel('胸痛类型')
plt.ylabel('计数')
plt.show()

X = data.drop('心脏病发作几率（0=更少，1=更多）', axis=1)
y = data['心脏病发作几率（0=更少，1=更多）']
rf = RandomForestClassifier()
rf.fit(X, y)

feat_importances = pd.Series(rf.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title('使用随机森林的特征重要性')
plt.xlabel('重要性')
plt.ylabel('特征')
plt.show()

pd.plotting.scatter_matrix(data[['患者年龄', '胆固醇（mg/dl）', '最大心率达到']], figsize=(10, 8))
plt.suptitle('散点矩阵图')
plt.show()

sns.pairplot(data[['患者年龄', '胆固醇（mg/dl）', '最大心率达到', '胸痛类型', '心脏病发作几率（0=更少，1=更多）']], kind='kde')
plt.title('核密度估计的成对关系图')
plt.show()

from pandas.plotting import parallel_coordinates

plt.figure(figsize=(10, 8))
parallel_coordinates(data[['患者年龄', '胆固醇（mg/dl）', '最大心率达到', '胸痛类型', '心脏病发作几率（0=更少，1=更多）']], '心脏病发作几率（0=更少，1=更多）')
plt.title('平行坐标图')
plt.show()


```
